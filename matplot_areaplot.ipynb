```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.plot(kind='area', stacked=False)
```

```python
my_df.abs().plot(kind='area')
```

```python
my_df.abs().plot(kind='area', alpha=0.5)
```

```python
my_df.abs().plot(kind='area', alpha=0.5, title='My awesome plot')
```

```python
my_df.abs().plot(kind='area', alpha=0.5, legend=False)
```

```python
my_df['Mon'].abs().plot(kind='area', alpha=0.5)
```

```python
my_df[['Mon','Thurs']].abs().plot(kind='area', alpha=0.5)
```

```python
my_df.abs().plot(kind='area', alpha=0.5, table=True)
```
