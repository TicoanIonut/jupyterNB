```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.abs().plot(kind="hexbin", x='Mon', y='Tues', gridsize=10)
```

```python
my_df.plot(kind="hexbin", x='Mon', y='Tues', gridsize=10, cmap="magma")
```

```python
my_df.plot.hexbin(x='Mon', y='Tues', C="Wed", gridsize=10)
```

```python
my_df.abs().plot(kind="hexbin", x='Mon', y='Tues', gridsize=10, grid=True, cmap="magma")
```
