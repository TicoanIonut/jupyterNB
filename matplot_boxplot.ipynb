```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.plot(kind='box', figsize=(10, 5), cmap='jet')
```

```python
my_df.abs().plot(kind='box', figsize=(10, 5), cmap='jet')
```

```python
my_df.abs().plot(kind='box', figsize=(10, 5), cmap='jet', grid=True)
```
