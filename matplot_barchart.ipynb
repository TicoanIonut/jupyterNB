```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.abs().plot(kind='bar')
```

```python
my_df.abs().plot(kind='bar', stacked=True)
```

```python
my_df.abs().plot(kind='bar', stacked=True, title='Awesome Barchart')
```

```python
my_df.abs().plot.bar(stacked=True)
```
