```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.abs().plot(kind='scatter', x='Mon', y='Tues', c='Wed')
```

```python
my_df.plot(kind='scatter', x='Mon', y='Tues', c='Wed', cmap='magma')
```

```python
my_df.abs().plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*100, alpha=0.3, cmap='magma')
```

```python
cmap = 'magma'
colors = my_df.abs()['Wed']
```

```python
my_df.abs().plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*100, alpha=0.3, cmap=cmap, c=colors)
```
