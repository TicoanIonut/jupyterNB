```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.plot(kind='line')
```

```python
my_df.abs().plot(kind='line', lw=3)
```

```python
my_df["Mon"].abs().plot(kind='line', lw=2)
```

```python
my_df[["Mon", 'Tues']].abs().plot(kind='line', lw=2)
```

```python
my_df[["Mon", 'Tues']].abs().plot(kind='line', lw=2, figsize=(10,5))
```

```python
my_df.abs().plot.line(figsize=(10,5))
```
